## Basic Configurarion

```python
import paramiko
import time

ssh = paramiko.SSHClient() #initializing the SSHClient method from paramiko class
ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy()) #setting up automatic click response to connection

ssh.connect(hostname='192.168.56.100',port=22,username='python',password='Huawei@123')

cli = ssh.invoke_shell() #invonking the shell
cli.send('N\n') #sending command imput into the shell
cli.send('screen-length 0 temporary\n')
cli.send("display current configuration\n")
time.sleep(3) ## 3 seconds delay to retrive input from the shell

dis_cu = cli.recv(999999).decode()
print(dis_cu)

ssh.close() # closing session
```
